###### d. Simple Linear Regression Model and Analysis

```python
LR=LinearRegression()
```

```python
X=df[["Durasi Tidur"]] 
y=df[["Berat Badan"]]

model=LR.fit(X,y) 
```

<font color=fanta>**Coefficient Determination - $R^2$** </font>

```python
R_sq=model.score(X,y)
print("The R Square is :",R_sq.round(4))
```

$R^2$ menunjukkan 0.16% variasi berat badan seseorang dapat dijelaskan oleh variabelitas durasi tidur. Angka ini sangat rendah hampir mendekati 0. Sehingga regression model ini tidak kuat dan tidak cocok untuk memperkirakan berat badan seseorang.

Selanjutnya kita memeriksa linear relationship strenght antara durasi tidur dengan berat badan dengan mencari coefficient correlation.

```python
R=(R_sq)**(1/2)
print("The R is :",R.round(4))
```

R = 0.0396 menunjukkan korelasi yang lemah antara durasi tidur dengan berat badan.

<font color=fanta>**Intercept- $ùëè_0$** </font>

```python
b0=model.intercept_[0]
print( 'The intercept b0:', b0.round(4))
```

<font color=fanta>**Slope - $b_1$** </font>

```python
b1=model.coef_[0][0]
print('The slope b1:', b1.round(4))
```

Jadi, estimated regression modelnya adalah y = 49.1204 + 0.1486x.

Intercept $ùëè_0$ = 49.1204 menunjukkan bahwa ketika rata-rata durasi tidur seseorang adalah 0 jam maka berat badannya adalah 49.1204.
Slope $b_1$ = 0.1486 menunjukkan bahwa rata-rata durasi tidur yang bertambah setiap jamnya,berat badannya juga akan bertambah sebesar 0.1486 kg.

##### <font color=blue> **Predict Responses** </font>

```python
C=model.predict(X)
df["Estimated Weight"]=C
df
```

```python
df['Residual']=df['Berat Badan']-df['Estimated Weight']
df
```

```python
np.sum(df['Residual'])
```

```python
df["Residual Square"]=df["Residual"]**2
df
```

```python
np.sum(df["Residual Square"])
```

```python
X=sm.add_constant(df["Durasi Tidur"].to_numpy())
y=df["Berat Badan"].to_numpy()
LR=sm.OLS(y,X)
model=LR.fit()
print(model.summary())
```
